### Task-which raw material is used maximum?

```python
dataset.ingredients
```

```python
dataset.ingredients[0]
```

```python
material_list=[]
for material_string in dataset.ingredients:
    material_string=material_string.lower()
    materials=material_string.split(",")
    materials=[material.strip() for material in materials]
    for material in materials:
        if material not in material_list:
            material_list.append(material)
print("There are {} unique elements in the dataset".format(len(material_list)))            
```

```python
print(material_list)
```

creating a surpass matrix.

```python
matrix_list=[]
for material in material_list:
    column=[]
    for ingredient in dataset.ingredients:
        ingredient=ingredient.lower().replace(",",' ')
        if material in ingredient:
            column.append(1)
        else :
            column.append(0)
    matrix_list.append(column)        
```

```python
matrix_array=np.array(matrix_list)
```

```python
matrix_array
```

```python
print("shape of matrix is {}".format(matrix_array.shape))
```

```python
material_dataframe=pd.DataFrame(matrix_array.T)
material_dataframe.columns=material_list
material_dataframe.head()
```

```python
material_lists=material_dataframe.sum(axis=0).to_list()
```

```python
material_lists
```

```python
len(material_lists)
```

```python
plt.figure(figsize=(28,8))
sns.barplot(x=material_list,y=material_lists)
plt.xticks(size=6,rotation='vertical')
plt.show()
```
